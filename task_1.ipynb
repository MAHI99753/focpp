```python
print("Beckett Pizza Plaza 4-for-3 Offer")
print("=================================")

prices = []
pizza_number = 1

# Keep asking until we have 4 valid prices
while len(prices) < 4:
    try:
        price = float(input(f"Enter Price of Pizza #{pizza_number}: "))
        
        if price <= 0:
            print("Please enter a valid price!")
            continue
        
        prices.append(price)
        pizza_number += 1

    except ValueError:
        print("Please enter a valid price!")

# Calculate totals
original_total = sum(prices)
discount = min(prices)          # Cheapest pizza is free
final_total = original_total - discount
discount_percent = (discount / original_total) * 100

# Display result
print()
print(f"Order Total is £{final_total:.2f}, "
      f"a fabulous discount of {discount_percent:.0f}%!")
```
