```python
#1.Write a function that accepts a positive integer as a parameter and then returns a
#representation of that number in binary (base 2).
def decimal_to_binary(n):
    binary = ""
    while n > 0:
        remainder = n % 2
        binary = str(remainder) + binary
        n = n // 2
    return binary
num = int(input("Enter a positive integer: "))
print("Binary representation:", decimal_to_binary(num))

```

```python
#2.Write and test a function that takes an integer as its parameter and returns the
#factors of that integer. (A factor is an integer which can be multiplied by another to
#yield the original). 
def find_factor(n):
    factor = []
    for i in range(1, n + 1):
        if n % i == 0:
            factor.append(i)
    return factor

num = int(input("Enter an integer: "))
print("Factors are:", find_factor(num))
```

```python
#3.Write and test a function that determines if a given integer is a prime number. A
#prime number is an integer greater than 1 that cannot be produced by multiplying
#two other integers. 
def find_prime(num):
    if num <= 1:
        return "Not a prime number"

    for i in range(2, num):
        if num % i == 0:
            return "Not a prime number"

    return "Prime number"


prime_number = int(input("Enter the number you want to check: "))
print("the number is::::",find_prime(prime_number))
```

```python
#4.Computers are commonly used in encryption. A very simple form of encryption
#(more accurately "obfuscation") would be to remove the spaces from a message
#and reverse the resulting string. Write, and test, a function that takes a string
#containing a message and "encrypts" it in this way. 
def encrypt_msg(message,shift): 
    encrypted=""
    for char in message: 
        if char.isalpha():
            encrypted+=chr(ord(char)+shift) 
        else: 
            encrypted+=char
    return encrypted 
msg=input("enter a message:") 
key=int(input("enter the shift value"))
print("encrypted message:",encrypt_msg(msg,key))
    
```

```python
#5.Another way to hide a message is to include the letters that make it up within seemingly random text. 
#The letters of the message might be every fi h character,
#for example. Write and test a function that does such encryption. It should
#randomly generate an interval (between 2 and 20), space the message out
#accordingly, and should fill the gaps with random letters. The function should
#return the encrypted message and the interval used.
#For example, if the message is "send cheese", the random interval is 2, and for
#clarity the random letters are not random:send cheese
#s  e  n  d   c  h  e  e  s  e
#sxyexynxydxy cxyhxyexyexysxye 
import random

def encrypt_message(message):
    interval = random.randint(2, 20)
    encrypted = ""

    for ch in message:
        encrypted += ch
        for _ in range(interval - 1):
            encrypted += chr(random.randint(97, 122)) # ASCII value

    return encrypted, interval

msg = "send cheese"

encrypted, gap = encrypt_message(msg)

print("Original :", msg)
print("Encrypted:", encrypted)
print("Interval :", gap)

```

```python
#Write a program that decrypts messages encoded as above 
msg = "send cheese"

# Encryption with fixed interval 
interval = 3
encrypted = ""

for ch in msg:
    encrypted += ch
    for _ in range(interval - 1):
        encrypted += "x"  # using 'x' instead of random letters for simplicity

print("Encrypted:", encrypted)

# Decrypt
def decrypt_message(encrypted, interval):
    return encrypted[::interval]

original = decrypt_message(encrypted, interval)
print("Decrypted:", original)
```
