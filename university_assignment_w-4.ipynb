```python
#1.Functions are often used to validate input. Write afunction that accepts a single
#integer as a parameter and returns Trueif the integer is in the range 0 to 100
#(inclusive), or False otherwise. Write a short program to test the function.  
def input_user():
    user_input = int(input("Enter a number between 0 and 100: "))
    
    if user_input >= 0 and user_input <= 100:
        return True
    else:
        return False

print(input_user())
```

```python
#2.Write a function that has a single string as its parameter, and returns the number of
#uppercase letters, and the number of lowercase letters in the string. Test the
#function with a short program 
def input_user():
    user_input = input("Enter your text : ")
    count_upper = 0
    count_lower = 0
    for ch in user_input:
        if ch.isupper():
            count_upper += 1
        elif ch.islower():
            count_lower += 1
    print("The number of uppercase letters :", count_upper)
    print("The number of lowercase letters :", count_lower)

input_user()
```
